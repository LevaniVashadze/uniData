```python
import pandas as pd
```

```python
data = pd.read_excel("grants.xlsx")
```

```python
filtered_data = data[data["აკად/მოსამზად"] == "აკად"]
filtered_data["გრანტი %"].fillna(0, inplace = True)
```

```python
filtered_data.groupby(["უსდ კოდი", "უსდ"])["გრანტი %"].mean()
```

```python
filtered_data.groupby(["უსდ", "პროგ. კოდი","პროგრამა"])["გრანტი %"].mean()
```
